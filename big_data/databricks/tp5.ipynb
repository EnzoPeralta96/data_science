```python
# 1 - Lectura de datos
df_vuelos = spark.read.csv("/Workspace/Users/enzash.kiv@gmail.com/airlines_a.csv", header=True, inferSchema=True)

# Analizar columnas, cantidad de datos y tipos
print("Vuelos ----------")
df_vuelos.printSchema()
print("Cantidad de filas:", df_vuelos.count())
print("Cantidad de columnas:", len(df_vuelos.columns))
print("Columnas:", df_vuelos.columns)
```

```python
# Lectura de datos
df_empresas = spark.read.csv("/Workspace/Users/enzash.kiv@gmail.com/L_UNIQUE_CARRIERS.csv", header=True, inferSchema=True)

# Analizar columnas, cantidad de datos y tipos
print("Empresas ----------")
df_empresas.printSchema()
print("Cantidad de filas:", df_empresas.count())
print("Cantidad de columnas:", len(df_empresas.columns))
print("Columnas:", df_empresas.columns)

df_empresas.show(5)
```

```python
#Daraframe filtrado por las columnas indicadas.
columns =  ['OP_CARRIER', 'ORIGIN', 'DEST', 'DEP_TIME', 'DEP_DELAY',
            'ARR_TIME', 'ARR_DELAY', 'DISTANCE', 'FL_DATE']

df_vuelos_reducido = df_vuelos.select(columns)

display(df_vuelos_empresas.limit(5))
```

```python
# Unir vuelos con empresas
df_vuelos_empresas = df_vuelos_reducido.join(
    df_empresas,
    df_vuelos_reducido.OP_CARRIER == df_empresas.Code,
    'inner'
)

display(df_vuelos_empresas.limit(20))
```

```python
# Cantidad de vuelos de cada una de las empresas
cantidad = df_vuelos_empresas.groupBy('Description').count()
display(cantidad)
```

```python
#Calcular la demora de cada vuevlo 
#Suma de los delay de salida y llegada
from pyspark.sql.functions import col
df_vuelos_empresas = df_vuelos_empresas.withColumn("Demora", col("DEP_DELAY") + col("ARR_DELAY") )

display(df_vuelos_empresas.limit(20))

```

```python
# Promedio de demora de vuelvo cada empresa 
promedio = df_vuelos_empresas.groupBy("Description").avg("Demora")
display(promedio)
```

```python
# Plot de demora por empresa
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff

fig = px.bar(promedio, x="Description", y="avg(Demora)", title="Promedio de demora por empresa")
fig.show()
```
