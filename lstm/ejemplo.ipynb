```python
import numpy as np
import matplotlib.pyplot as plt

# --- Datos de ejemplo ---
test_scaled = np.array([1, 2, 3, 4, 5, 6, 7, 8])
n_input = 3
n_output = 2

# Crear ventanas deslizantes (4 posibles)
windows = [test_scaled[i : i + n_input] for i in range(len(test_scaled) - n_input - n_output + 1)]
windows = np.array(windows)

# Simular predicciones del modelo (solo para el ejemplo)
preds = np.array([
    [4.1, 5.1],
    [5.0, 6.1],
    [6.1, 7.0],
    [7.0, 8.1]
])

# --- Grfico ---
plt.figure(figsize=(10, 5))
plt.plot(test_scaled, 'o-', color='gray', label='Serie original (SYM-H)')
plt.grid(True, linestyle='--', alpha=0.5)

# Colores distintos para cada ventana
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd']

for i, w in enumerate(windows):
    # Dibujar la ventana de entrada
    plt.plot(range(i, i + n_input), w, 'o-', color=colors[i], label=f'Ventana {i+1} (entrada)')
    # Dibujar las predicciones que siguen
    plt.plot(range(i + n_input, i + n_input + n_output), preds[i], 'x--', color=colors[i], label=f'Prediccin {i+1}')

plt.title('Ejemplo visual de Sliding Window + Prediccin Multi-Step')
plt.xlabel('ndice temporal (t)')
plt.ylabel('Valor (SYM-H escalado)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
