## **1. ¿Qué es Pandas?**

image.png

Python ofrece funciones integradas para crear, escribir y leer archivos. Hay dos tipos de archivos que se pueden manejar en Python: archivos de texto normales y archivos binarios.

**En este tutorial, en lugar de las funciones integradas de Python, trabajaremos con una potente herramienta para gestionar archivos: Pandas**

**Pandas** es una biblioteca/paquete de Python que permite manipular datos de una forma muy sencilla y eficiente. Su objetivo es ser una herramienta flexible para el análisis de datos de alto rendimiento.

Algunas características:

- Maneja dos tipos fundamentales de estructuras de datos: DataFrame y DataSeries.
- Estructura DataFrame rápida y eficiente con indexación predeterminada y personalizada.
- Herramientas para cargar datos desde diferentes formatos de archivo.
- Manejo integrado de datos faltantes.
- Segmentación, indexación y subconjunto de grandes conjuntos de datos basados ​​en etiquetas.
- Capacidad para eliminar o insertar columnas de una estructura de datos.
- Agregación y transformaciones de datos.
- Fusión y unión de datos de alto rendimiento.
- Funciones para el preprocesamiento de series temporales.

###**4. Slicing datasets (loc, iloc)**

Pandas es muy útil para la manipulación de datos matemáticos y se utiliza ampliamente en el campo del aprendizaje automático. Incluye muchos métodos para su correcto funcionamiento. **loc()** e **iloc()** son uno de esos métodos. Se utilizan para cortar datos del DataFrame de Pandas. Ayudan en la selección conveniente de datos del DataFrame. Se utilizan para filtrar los datos según algunas condiciones.

- loc: La función loc() es un método de selección de datos **basado en etiquetas**, lo que significa que tenemos que pasar el nombre de la fila o columna que queremos seleccionar.

- iloc: La función iloc() es un método de selección **basado en índices**, lo que significa que tenemos que pasar un índice entero en el método para seleccionar una fila/columna específica.

Diferencias entre **loc** e **iloc**:
- Loc() incluye el último elemento del rango que se le pasa, a diferencia de iloc().
- loc() puede aceptar datos booleanos a diferencia de iloc().

###**5. Cleaning data (corrección de formatos de datos, reemplazar valores, eliminar filas, eliminar columnas, eliminar duplicados)**

```python
import pandas as pd

#crear un dataframe manualente con datos que permitan mostrar cleaning data
datos = [["juan", "salguero", "", "estudiate", "1986-01-02"], ["juan", "salguero", "", "estudiate", "1986-01-02"], ["maria", "estefan", "33", "profesora", "1990-09-15"],["constanza", "pinto", 13, "estudiate", "2000-10-12"],["julian", "ponce", 29, "estudiate", "1994-04-20"]]
columnas = ["nombre", "apellido", "edad", "ocupacion", "fecha_nacimiento"]

df1 = pd.DataFrame(datos, columns=columnas)
df1.to_csv("datos_para_test.csv")

df2 = pd.read_csv("datos_para_test.csv", index_col=0)
print(df2)
print("\n\n")
print(df2.info())
```

```python
#corrección de formato de datos.

#es posible hacer una diferencia de fechas entre los campos?
print(type(df2["fecha_nacimiento"].iloc[0]))
fecha1 = df2["fecha_nacimiento"].iloc[0]
fecha2 = df2["fecha_nacimiento"].iloc[2]
print(fecha1, fecha2)
#fecha3 = fecha1-fecha2 #aquí dará error

def parser_de_string_en_anio_mes_dia(cadena):
  cadena = cadena.split("-")
  return cadena[0], cadena[1], cadena[2]

anioFecha1, mesFecha1, diaFecha1 = parser_de_string_en_anio_mes_dia(fecha1)
anioFecha2, mesFecha2, diaFecha2 = parser_de_string_en_anio_mes_dia(fecha2)
print(anioFecha1, mesFecha1, diaFecha1)

from datetime import datetime
fecha1dt = datetime(int(anioFecha1), int(mesFecha1), int(diaFecha1))
fecha2dt = datetime(int(anioFecha2), int(mesFecha2), int(diaFecha2))

fecha3 = fecha2dt - fecha1dt
print(fecha3)
print("\n\n")
#Se corrige el formato del campo fecha_nacimiento, convirtiédolo a datetime
df2['fecha_nacimiento'] = pd.to_datetime(df2['fecha_nacimiento'])
print(df2.info())
```

```python
#reemplazar valores. Detectar valores nulos. Aprovechar el campo fecha_nacimiento para completar edades faltantes

#contar valores nulos de la columna edad. Mostrar cuáles son las filas con valores nulos. imputar valor edad con edad calculada usando columna fecha_nacimiento.
#contar valores nulos de la columna edad
nulosColumnaEdad = df2['edad'].isnull().sum()
print(nulosColumnaEdad)
print("\n---\n")

#mostrar cuáles son las filas con valores nulos.
filasConNulosColumnaEdad = df2[df2['edad'].isnull()]
print(filasConNulosColumnaEdad)
print("\n---\n")

#imputar con valor constante
'''
df2.loc[df2['edad'].isnull(), 'edad'] = 5
print(df2)
print("\n---\n")
'''

#imputar valor edad con edad calculada usando columna fecha_nacimiento
def calcular_edad(fecha_nacimiento):
    """
    Calcula la edad de una persona basada en su fecha de nacimiento.
    :param fecha_nacimiento: datetime.date, la fecha de nacimiento.
    :return: int, la edad de la persona en años.
    """
    hoy = datetime.today()
    edad = hoy.year - fecha_nacimiento.year
    # Ajustar si la fecha de cumpleaños aún no ha ocurrido este año
    if (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day):
        edad -= 1
    return edad

df2['edad'] = df2['fecha_nacimiento'].apply(calcular_edad)
print(df2)
print("\n---\n")

#y si solamente querría imputar las edades de las filas faltantes y no recorrer toda la data?
#df2[df2['edad'].isnull()]["edad"] = 5  #esto es intuitivo, pero no funcona. Error: Try using .loc[row_indexer,col_indexer] = value instead  --> Explicar
```

```python
#eliminar filas
#eliminar filas que posee campo edad en NaN
print(df2)
print("\n---\n")
filasConNulosColumnaEdad = df2[df2['edad'].isnull()]
indicesConNulos = filasConNulosColumnaEdad.index.tolist()
print(indicesConNulos)
df2 = df2.drop(indicesConNulos)
print(df2)
print("\n---\n")
```

```python
#eliminar duplicados
print(df2)
print("\n---\n")
df2 = df2.drop_duplicates()
print(df2)
```

```python
#eliminar columnas
print(df2)
print("\n---\n")
df2 = df2.drop(columns=['edad'])
print(df2)
```

###**7. Filtrar datos (DataFrame.query())**
Pandas tiene el método query, que permite realizar consultas más legibles cuando las condiciones son complejas.

```python
#crear un dataframe manualente con datos que permitan mostrar cleaning data
datos = [["maria", "albarado", "40", "supervisor", "1984-05-17"], ["juan", "salguero", "38", "estudiante", "1986-01-02"], ["juan", "salguero", "38", "estudiante", "1986-01-02"], ["maria", "estefan", "33", "docente", "1990-09-15"],["constanza", "pinto", 13, "estudiante", "2000-10-12"],["julian", "ponce", 29, "estudiante", "1994-04-20"]]
columnas = ["nombre", "apellido", "edad", "ocupacion", "fecha_nacimiento"]
df2 = pd.DataFrame(datos, columns=columnas)
print(df2.info())
print("\n---\n")
print(df2)
```

```python


#sin el casteo, puede fallar la query
df2["edad"] = df2["edad"].astype(int)
print(df2.info())

resultado = df2[df2['edad'] > 30]
print(resultado)
print("\n\n")
resultado = df2.query('edad > 30')
print(resultado)
```

###**8. Operaciones entre Dataframes: join (left outer join, right outer join, inner join), concat**


image.png

```python
import pandas as pd

# DataFrame 1
data1 = {'ID': [1, 2, 3, 4],
         'Nombre': ['Juan', 'María', 'Pedro', 'Lucía']}
df1 = pd.DataFrame(data1)

# DataFrame 2
data2 = {'ID': [3, 4, 5, 6],
         'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']}
df2 = pd.DataFrame(data2)

print(df1)
print("\n\n", df2)
```

```python
# Realizar un inner join
inner_join_result = pd.merge(df1, df2, on='ID', how='inner')
print(inner_join_result)
print("\n---\n")

# Realizar un left join
left_join_result = pd.merge(df1, df2, on='ID', how='left')
print(left_join_result)
print("\n---\n")

# Realizar un right join
right_join_result = pd.merge(df1, df2, on='ID', how='right')
print(right_join_result)
print("\n---\n")

# Realizar un full outer join
full_outer_join_result = pd.merge(df1, df2, on='ID', how='outer')
print(full_outer_join_result)
```

```python
# Crear dos DataFrames de ejemplo
data1 = {'ID': [1, 2, 3],
         'Nombre': ['Juan', 'María', 'Pedro']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [4, 5, 6],
         'Nombre': ['Lucía', 'Ana', 'Carlos'],
         'Apellido': ['Juarez', 'Jerez', 'Pinto']
         }
df2 = pd.DataFrame(data2)

print(df1)
print("\n\n {} \n\n".format(df2))
# Concatenar los DataFrames por filas
df_concat = pd.concat([df1, df2], axis=0)
print(df_concat)
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo: Años y temperaturas medias
años = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
temperaturas = [14.5, 14.8, 14.7, 15.0, 15.2, 15.5, 15.7, 15.8, 16.0, 16.2, 16.3]

# Crear el gráfico de línea
plt.figure(figsize=(10, 6))  # Opcional: para ajustar el tamaño de la figura
plt.plot(años, temperaturas, marker='o', linestyle='-', color='b', label='Temperatura Media')

# Agregar título y etiquetas de ejes
plt.title('Temperatura Media Anual (2010-2020)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)

# Agregar una cuadrícula
plt.grid(True)

# Agregar una leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
