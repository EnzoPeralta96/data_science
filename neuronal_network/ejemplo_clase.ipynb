```python
import pandas as pd
```

```python
df = pd.read_csv("sample_data/ionosphere.data", header= None)

print(df)
```

```python
df.describe()
```

Renombramos primera fila de la forma : f1_real, f1_img, f2_real, f2_img...

```python
colums = list()

for i in range(1,18):
    col1 = "f" + str(i) + "_r"
    col2 = "f" + str(i) + "_i"
    colums.append(col1)
    colums.append(col2)

colums.append("target")

df.columns = colums

print(df)
```

Convertir el target de g / b a 1 y 0 respectivamente

```python
clases = {'g' : 1, 'b': 0}
df["target"] = df["target"].map(clases)

print(df)
```

```python
#Ingeneria de datos - analisis de balance de datos.
print(df["target"].value_counts())
```

```python
#balancear datos:
'''
    Como hay mas datos con buenas medicion que malas, equilibro el set de datos en la misma
    cantidad de buenas y malas.
    Esto se hace para evitar de introducir sesgo en el modelo.
    Ya que al tener mayor casos de los buenos, el modelo aprendera mas de los buenos.

    Otra estrategia puede ser rellenar con valores malos -> entonces estos son datos sinteticos
    que pueden introducir ruido en el modelo.
    Hay que tener un criterio para eso.
    Ver ventajas y desventajas de esto.
'''
#Creo un nuevo dataframe con la misma cantidad de datos.
# Pandas no funciona bien con un dataframe muy grande, para iterar sobre filas hay
# que usar rows, que esta optimizado para eso.

contador = 99
df_nuevo = pd.DataFrame()

for index, rows in df.iterrows():
    if contador > 0:
        if rows['target'] == 1:
            contador -= 1
        else:
            df_nuevo = pd.concat([df_nuevo, pd.DataFrame([rows])])
    else:
        df_nuevo = pd.concat([df_nuevo, pd.DataFrame([rows])])

```

```python
print(df_nuevo["target"].value_counts())
```

```python
#dividir caracteristicas y target:
features = df_nuevo.loc[:, df_nuevo.columns != 'target']
target = df_nuevo.loc[:, df_nuevo.columns == 'target']
```

```python
#dividir los datos en entrenamiento y testeo. 
# Puede ser 80/20, 70/30
# Tambien depende de los datos que tengamos.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#crear el modelo con sequential, dense, activation.
# instalar tensorflow (incluye tensorflow.keras) si falta

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
# use number of features as input dimension
imput_dim1 = x_train.shape[1]
print(imput_dim1)
modelo = Sequential()
# specify input_dim (integer) and use a valid activation
modelo.add(Dense(64, input_dim=imput_dim1, activation='relu'))
# output layer for binary classification with sigmoid
modelo.add(Dense(1, activation='sigmoid'))

# fix metrics name
modelo.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# pass labels as 1D array
historial = modelo.fit(x_train, y_train.values.ravel(), epochs=10, batch_size=32, validation_split=0.2, verbose=2)
```

```python
#Graficar luego loss vs epoch
#Graficar accuracy vs epoc
```

```python
y_pred = modelo.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
y_pred_binario = (y_pred > 0.7 ).astype(int)
```

```python
matriz_conf = confusion_matrix(y_test, y_pred_binario)
#En la diagonal principal True_negative y true_positivo
#Ver documentacion de la teoria.
print(matriz_conf)
```
